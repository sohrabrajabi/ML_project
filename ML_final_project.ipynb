## Predict whether a mammogram mass is benign or malignant

```python
# import libraries

import sys
import os
import pandas as pd
```

```python
# data preperation and wrangling
masses_data = pd.read_csv('mammographic_masses.data.txt')
masses_data.head()
```

```python
# update features names
# assighn NaN value to question marks
masses_data = pd.read_csv('mammographic_masses.data.txt', na_values = ['?'],
                          names = ['BI-RADS', 'age', 'shape', 'margin', 'density', 'severity'])   
masses_data.head()
```

```python
masses_data.describe()
```

Number of data which is available for density is widely different with other variables. Befor we just drop every row with missing dat, let's make sure we don't bias our data by doing so. We should check does there appear to be any sort of correlation to what sort of data missing fields? If there were, we would have to try and go back and fill that data in.

```python
# pull up any row which contains missing data in either the four feature columns 
masses_data.loc[(masses_data['age'].isnull())|
                (masses_data['shape'].isnull())|
                (masses_data['margin'].isnull())|
                (masses_data['density'].isnull())]
```

data seems randomly distributed, go ahead and drop rows with missing data.

```python
masses_data.dropna(inplace = True)
masses_data.describe()
```

Now we can convert the Pandas dataframe to into numpy arrays that can be used by scikit_learn. Here we need for features of; age, shape, margin, density and another array to contains the classes(severity). We also need an array of the features name labels.

```python
all_features = masses_data[['age', 'shape', 'margin', 'density']].values

all_classes = masses_data['severity'].values

feature_names = ['age', 'shape', 'margin', 'density']

all_features
```

Some of our models require the input data to be normalized. by preprocessing

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
all_features_scaled = scaler.fit_transform(all_features)
all_features_scaled
```

```python
# lets start with 75% of data for training and 25% for testing
import numpy
from sklearn.model_selection import train_test_split

numpy.random.seed(1234)

(training_inputs, testing_inputs,
 training_classes, testing_classes) = train_test_split(all_features_scaled, all_classes,
                                                       train_size = 0.75, random_state = 1)
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state = 1)

clf.fit(training_inputs, training_classes)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf, filled=True)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, testing_inputs, testing_classes)
print("%0.5f accuracy with a standard deviation of %0.5f" % (scores.mean(), scores.std()))
```

Know instead of a single train/test split, use K-Fold cross validation to get a better measure of our model's accuracy (K=10).

```python
from sklearn.model_selection import cross_val_score

clf = DecisionTreeClassifier(random_state = 1)

cv_score = cross_val_score(clf, all_features_scaled, all_classes, cv = 10)

print("%0.5f accuracy with a standard deviation of %0.5f" % (cv_score.mean(), cv_score.std()))
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 10, random_state = 1)

cv_score = cross_val_score(clf, all_features_scaled, all_classes, cv = 10)

print("%0.5f accuracy with a standard deviation of %0.5f" % (cv_score.mean(), cv_score.std()))
```

## SVM
